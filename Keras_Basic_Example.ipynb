```python
from keras.modelsdels import Sequential
from keras.layers import Dense
import numpy as np
```

```python
# Create a model
model = Sequential()
```

```python
# Stack Layers
model.add(Dense(units = 64, activation='relu', input_dim = 100))
model.add(Dense(units = 10, activation='softmax'))
```

```python
# Configure leaning
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
# Create Numpy arrays with random values, use your training or test data here
train_data = np.random.random((64,100))
result_data = np.random.random((64,10))
test_train_data = np.random.random((64,100))
test_result_data = np.random.random((64,10))
```

```python
result_data
```

```python
# Train using numpy arrays
model.fit(train_data, result_data, epochs =5, batch_size = 32)
```

```python
loss_and_metrics = model.evaluate(test_train_data, test_result_data,batch_size=128)
```

```python
classes = model.predict(test_train_data, batch_size = 128)
classes
```
